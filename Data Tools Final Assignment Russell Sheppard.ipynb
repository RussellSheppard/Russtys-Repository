```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy
import seaborn as sns
from collections import Counter
```

```python
# Load Data
df = pd.read_csv("survey.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
df.dropna()
```

```python
df.index
```

```python
#Changing a categorical to numeric
df["treatment"] = df["treatment"].replace(["Yes", "No"], [1, 0])
```

```python
df.treatment
```

```python
# "Have you sought treatment for a mental health condition?"
df.treatment
```

```python
# Do you have a family history of mental illness?
# Changing the categorical to a numeric
df["family_history"] = df["family_history"].replace(["Yes", "No"], [1, 0])
```

```python
df.family_history
```

```python
# Does your employer provide resources to learn more about mental health issues and how to seek help?
# Changing categorical to numeric
df["seek_help"] = df["seek_help"].replace(["Yes", "Don't know", "No"], [1, 0.5, 0])
```

```python
df.seek_help
```

```python
# Do you feel that your employer takes mental health as seriously as physical health?
# Changing the categorical to a numeric
df["mental_vs_physical"] = df["mental_vs_physical"].replace(["Yes", "Don't know", "No"], [1, 0.5, 0])
```

```python
df.mental_vs_physical
```

```python
test1=df.groupby(["Country"])["treatment"].count().reset_index().sort_values("treatment", ascending=False)
print(test1)
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.bar(x= test1.Country, height=test1.treatment)
# Add Chart Characteristics
plt.title("Mental Health Frequency")
plt.xticks(rotation="vertical")
plt.xlabel("Country")
plt.ylabel("Treatment")
plt.tight_layout()
fig.savefig("mentalhealthfrequency.png")
```
